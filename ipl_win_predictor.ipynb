```python
delivery_df["current_score"] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df["runleft"] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df['ball_left']=120-(delivery_df['over']*6+ delivery_df['ball'])
```

```python
# Step 1: Convert dismissed to 0/1 properly
delivery_df['is_dismissed'] = delivery_df['player_dismissed'].notna().astype(int)

# Step 2: Group by match and do cumulative sum
delivery_df['wicket'] = 10 - delivery_df.groupby('match_id')['is_dismissed'].cumsum()
delivery_df.tail()
```

```python
delivery_df['crr']= (delivery_df['current_score']*6)/(120-delivery_df['ball_left'])
delivery_df.head()
```

```python
delivery_df['rrr']= (delivery_df['runleft']*6)/(delivery_df['ball_left'])
delivery_df.head()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
  
```

```python
  delivery_df['target']= delivery_df.apply(result, axis=1)
```

```python
final_df=delivery_df[['batting_team','bowling_team','city','runleft','ball_left','wicket','total_runs_x','crr','rrr','target']]
```

```python
final_df.head()
```

```python
final_df= final_df.sample(final_df.shape[0])
```

```python
final_df.isnull().sum()
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df['ball_left'] !=0] 
```

##MODEL BUILDING

```python
X = final_df.iloc[:, :-1]
Y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([('trf',OneHotEncoder(sparse_output=False, drop="first"),['batting_team','bowling_team','city'])
                        ], remainder="passthrough")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ("step1",trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred= pipe.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```
